```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# OPTIONAL (more relevant for individual words)
from nltk.stem import LancasterStemmer, WordNetLemmatizer 
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import re   # regex
import unicodedata
import nltk
# NOT NEEDED import json
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../raw/Onboard_Survey.csv")

df.head()
```

```python
# skip first seven columns df.iloc[:, 0:7].head()

# selecting only open-ended responses 
df.iloc[:, 6:14].head()

open_ended = df.iloc[:, 6:14]

open_ended.head()
```

```python
# Column: If you have used a wallet, what was your first wallet and why did you choose that one?

empty_list = []

 # store strings from each row in an empty list
for row in open_ended.iloc[:,0]:
    empty_list.append(row)

empty_list2 = []

# store first word from each string into empty_list2
# try-except to handle exception
# n = 180
for string in empty_list:
    try: 
        empty_list2.append(string.split()[0])
    except:
        continue
    
```

```python
# see how many strings out of 180 is stored
# we got 176
# NOTE: most people respond directly, but some answer in sentences

len(empty_list2)
```

```python
empty_list3 = []

# convert all words to lowercase and remove non-alphanumeric characters
for word in empty_list2:
    s = re.sub(r'\W+', '', word.lower())
    empty_list3.append(s)
    
pd.Series(empty_list3).value_counts()
```

```python
# split string, everything EXCEPT first word

empty_list2a = []

for string in empty_list:
    try:
        other = string.split()[1:]
    except:
        continue
    empty_list2a.append(' '.join(other))
    
empty_list2a
```

```python
len(empty_list2a)
```

```python
# rename columns
open_ended.columns = ['walletwhat_walletwhy', 'wallet_pain', 'defi_when', 'defiwhat_defiwhy', 'defi_pain', 'defi_outcome', 'defi_interest', 'defi_endgame']


open_ended.head()
```

```python
open_ended["walletwhat"] = open_ended["walletwhat_walletwhy"].str.split(' ').str[0]

open_ended.head()
```

```python
open_ended["walletwhat"] = open_ended["walletwhat"].str.lower()

open_ended.head()
```

```python
open_ended['defiwhat'] = open_ended["defiwhat_defiwhy"].str.split(' ').str[0]

open_ended.head()
```

```python
open_ended['defiwhat'] = open_ended['defiwhat'].str.lower()

open_ended.head()
```

```python
# first rename columns to make them manageable
# now that we've split 'wallet' and 'defi' into 'what' and 'why'
# we can clean the 'what' and tally
# we can filter by popular 'what' and quickly get a sense for the 'why' 
# **Filter by words should yield most impact in minimal time

# process the data: 
# (advanced w/ more time: we can apply topic modeling or LDA to the 'why')
```
